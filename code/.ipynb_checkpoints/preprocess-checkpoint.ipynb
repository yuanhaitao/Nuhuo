```python
# 构造新的edges：1.根据是否是oneway（单行）进行复制，2.根据geometry把一条edge切割成多条，每条的长度根据占比来计算
# 生成两个新文件，第一个是新的edges文件，主要保留id，长度，highway类型，上游link数组，下游link数组；第二个是老edge对应到新edge的关系
old2new_edge_dict = {}
new2old_edge_dict = {}
new_idx = 0
for idx in range(len(edge_pd)):
    edge = edge_pd.iloc[idx]
    geometry=edge['geometry'].strip()
    left = geometry.find("(")
    right = geometry.find(")")
    points = geometry[left+1:right].split(",") if left!=-1 and right!=-1 else None
    old2new_edge_dict[idx] = []
    if left == -1 or right == -1 or points==None or len(points)<=2:
        old2new_edge_dict[idx].append(new_idx)
        new2old_edge_dict[new_idx] = [idx, edge['length'], 1.0, edge['highway']]
        new_idx += 1
    else:
        seg_num = len(points)-1
        seg_len_list = []
        # print(points)
        for j in range(seg_num):
            xy_s = list(map(float,points[j].strip().split(" ")))
            xy_e = list(map(float,points[j+1].strip().split(" ")))
            delta_len = round(float(str(geodesic((xy_s[1],xy_s[0]), (xy_e[1],xy_e[0]))).split(" ")[0])*1000,4)
            seg_len_list.append(delta_len)
        # seg_ratio_list  = [round(t/sum(seg_len_list),4) for t in seg_len_list]
        for seg in seg_len_list:
            old2new_edge_dict[idx].append(new_idx)
            new2old_edge_dict[new_idx] = [idx, seg, round(seg/sum(seg_len_list),4), edge['highway']]
            new_idx += 1
```

```python
len(old2new_edge_dict),len(new2old_edge_dict)
# old2new_edge_dict
```

```python
# old2new_edge_dict
# 根据old2new_edge_dict和raw_node_edgeneighbor_dict构造新的edge上下游关系
new_edge2edge_neighbor_dict = {}
for k,links in old2new_edge_dict.items():
    u,v = edge_pd.iloc[k]['u'], edge_pd.iloc[k]['v']
    u_edges = []
    v_edges = []
    inverse_edges = []
    for t in raw_node_edgeneighbor_dict[u][1]:
        if t != k:
            u_t, v_t = edge_pd.iloc[t]['u'], edge_pd.iloc[t]['v']
            if u_t == v and v_t == u:
                # print(t,old2new_edge_dict[t],k,links)
                if len(old2new_edge_dict[t])==len(links):
                    inverse_edges.append(old2new_edge_dict[t])
            else:
                u_edges.append(old2new_edge_dict[t][-1])
    for t in raw_node_edgeneighbor_dict[v][0]:
        if t!=k and edge_pd.iloc[t]['v'] != u:
            v_edges.append(old2new_edge_dict[t][0])
    #u关系
    for u_e in u_edges:
        tmp1 = new_edge2edge_neighbor_dict.get(links[0],[[],[]])
        tmp1[0].append(u_e)
        new_edge2edge_neighbor_dict[links[0]] = tmp1
        tmp2 = new_edge2edge_neighbor_dict.get(u_e,[[],[]])
        tmp2[1].append(links[0])
        new_edge2edge_neighbor_dict[u_e] = tmp2
    #inverse关系，只改变自己，不改变对方
    for i_e in inverse_edges:
        for l_idx in range(len(links)):
            cur = links[l_idx]
            tmp1 = new_edge2edge_neighbor_dict.get(cur,[[],[]])
            tmp1[0].append(i_e[len(links)-1-l_idx])
            tmp1[1].append(i_e[len(links)-1-l_idx])
            new_edge2edge_neighbor_dict[cur] = tmp1
    #自关系
    for l_idx in range(len(links)-1):
        cur = links[l_idx]
        nxt = links[l_idx+1]
        tmp1 = new_edge2edge_neighbor_dict.get(cur,[[],[]])
        tmp2 = new_edge2edge_neighbor_dict.get(nxt,[[],[]])
        tmp1[1].append(nxt)
        tmp2[0].append(cur)
        new_edge2edge_neighbor_dict[cur] = tmp1
        new_edge2edge_neighbor_dict[nxt] = tmp2
```

```python
import os
# import metis
import pymetis as metis
import pandas as pd
from geopy.distance import geodesic
import pickle
import queue
raw_data_path="/home/hatim/data/EventDetection/chengdu"
my_data_path="/home/hatim/data/DataCompletion/data/chengdu"
if not os.path.exists(my_data_path):
    os.mkdir(my_data_path)
old2new_edge_dict, new2old_edge_dict, new_edge2edge_neighbor_dict = pickle.load(open(my_data_path+"/new_edge_dict.pk","rb"))
print(len(old2new_edge_dict))
print(len(new2old_edge_dict))
print(len(new_edge2edge_neighbor_dict))
```

```python
def partition(adjacency, nparts, ids):
    try:
        (_, parts) = metis.part_graph(nparts = nparts, adjacency=adjacency)
    except Exception as err:
        # print("error")
        print(err)
        parts = [0 for _ in range(len(ids)//2)] + [1 for _ in range(len(ids)//2,len(ids))]
    assert len(parts) == len(ids)
    id2part_dict = {i:idx for i,idx in zip(ids,parts)}
    nodeidx_per_part = {}
    for idx,p in enumerate(parts):
        nodeidx = nodeidx_per_part.get(p,{})
        nodeidx[idx] = len(nodeidx)
        nodeidx_per_part[p] = nodeidx
    target_adjacency_list = [[None for _ in range(len(nodeidx_per_part[t]))] for t in range(len(nodeidx_per_part))]
    target_ids_list = [[None for _ in range(len(nodeidx_per_part[t]))] for t in range(len(nodeidx_per_part))]
    cut_num = 0
    for i, nxt in enumerate(adjacency):
        i_part = id2part_dict[ids[i]]
        new_i = nodeidx_per_part[i_part][i]
        new_nxt = []
        for n in nxt:
            n_part = id2part_dict[ids[n]]
            if n_part != i_part:
                cut_num += 1
            else:
                new_nxt.append(nodeidx_per_part[n_part][n])
        target_adjacency_list[i_part][new_i] = new_nxt
        target_ids_list[i_part][new_i] = ids[i]
    return cut_num, id2part_dict, target_adjacency_list, target_ids_list

def layer_partition(adjacency, nparts_list, ids):
    total_id2part_dict = {i:[0 for _ in range(len(nparts_list))] for i in ids}
    subgraph_pool = queue.Queue()
    subgraph_pool.put((adjacency,0,ids))
    total_cut_num = 0
    epoch = 0
    while subgraph_pool.empty() is False:
        adj, np_idx, ids = subgraph_pool.get()
        print(epoch, np_idx, subgraph_pool.qsize())
        epoch += 1
        np = nparts_list[np_idx]
        cut_num, id2part_dict, target_adjacency_list, target_ids_list = partition(adj, np, ids)
        total_cut_num += cut_num
        for k,v in id2part_dict.items():
            total_id2part_dict[k][np_idx] = v
        if np_idx+1 < len(nparts_list):
            for ta, ti in zip(target_adjacency_list,target_ids_list):
                subgraph_pool.put((ta, np_idx+1, ti))
    return total_cut_num, total_id2part_dict
```

# 制作训练集数据

```python
#通过脚本制作，可以python的并行
# -*- encoding:utf-8 -*-
# 构造样本，每条样本是采样的一个子图（空间上），并且每个时间戳考虑其前后各一个小时（时间上）
# 将【25%】的mask掉作为自监督标签，mask用规则填补【50%填补默认值（限速+0），50%填补最近时间片的数据】
def load_traffic(files):
    '''加载所有的traffic信息'''
    res = {}
    for file in files:
        for log in open(file):
            data = log.split("\t")
            data = [d.strip() for d in data]
            assert len(data) == 290
            linkid, day = data[:2]
            linkid = int(linkid)
            tmp = res.get(linkid,{})
            tmp[day] = data[2:]
            res[linkid] = tmp
    return res
```

```python
import datetime
import copy
# begin = datetime.date(2016,10,1)
# end = datetime.date(2016,11,30)
# end = datetime.date(2016,10,7)
def build_dataset(begin, end, time_slot_num = 6, slip_step = 1):
    # 1.先加载数据
    d = begin
    delta = datetime.timedelta(days=1)
    days = []
    while d <= end:
        date = d.strftime("%Y%m%d")
        days.append(date)
        d += delta
    traffic_files = [my_data_path+"/traffic_{}".format(day) for day in days]
    traffic_dict = load_traffic(traffic_files)
    #2.利用加载的link分层图+link属性信息进行样本的构建（首先是link子图的遍历，其次是时间的遍历）
    first_num = len(first_layer_edges)
    for first_i in range(first_num):
        s_edges = second_layer_edges[first_i]
        second_num = len(s_edges)
        # 该层所有节点构成一个空间样本，接下来是时间遍历（根据begin和end确定的时间区间，如果是空的，那么就用''补全）
        # train_data = [] #按照时间片（一小时时间片，平移，slip的步数有参数确定,默认是1）
        last_date = None
        last_slot_info = []
        d = begin
        while d <= end:
            date = d.strftime("%Y%m%d")
            # 获取全天的子图列表（如果是空的，就用‘’补齐）
            data_list = []
            for second_i in range(second_num):
                data_list.append(traffic_dict.get(new_part2id_dict[(first_i, second_i)],{}).get(date,['' for _ in range(288)]))
            # print(date,len(data_list))
            if last_date != None:
                #那么last_data的最后time_slot_num-1个时间片信息可以接到今天
                for _t in range(time_slot_num-1):
                    print(len(last_slot_info),last_slot_info[0])
                    pre_set = [info[_t:] for info in last_slot_info]
                    assert len(pre_set) == len(data_list)
                    nxt_set = [info[:_t+1] for info in data_list]
                    train_set = [pre+nxt for pre,nxt in zip(pre_set, nxt_set)]
                    print(len(train_set[0]),train_set[0])
                    train_set = ['|'.join([train_set[s_i][t_i] for s_i in range(second_num)])  for t_i in range(time_slot_num)]
                    train_set = '\t'.join([str(first_i), last_date, str(288-(time_slot_num-_t)+1)] + train_set)
                    print(train_set)
                    break
            for _t in range(288-time_slot_num+1):
                train_set = ['|'.join([data_list[s_i][_t+t_i] for s_i in range(second_num)])  for t_i in range(time_slot_num)]
                train_set = '\t'.join([str(first_i), date, str(_t)] + train_set)
                print(train_set)
                break
            last_date = date
            last_slot_info = [copy.deepcopy(data[288-time_slot_num+1:]) for data in data_list]
            d += delta
```

```python
from sklearn.preprocessing import label_binarize
# from torch.utils.data import Dataset
import tqdm
import torch
import random
import pickle
import datetime
import dgl
import os
import numpy as np
# random.seed(1000)
from torch.utils.data import IterableDataset, DataLoader
class MyTrafficDataset(IterableDataset):                                                 
    def __init__(self,file_path: str, edge_file:str, graph_file:str):                                                                          
        super(MyTrafficDataset).__init__()                                        
        self.file_path = file_path                                              
        self.time_slot_size, self.end  = self._get_file_info(file_path)                             
        self.start = 0
        _, self.new2old_edge_dict, _ = pickle.load(open(edge_file,"rb"))
        _, self.second_layer_edges, new_id2part_dict, _ = pickle.load(open(graph_file,"rb"))
        self.new_part2id_dict = {(v[0],v[1]):k for k,v in new_id2part_dict.items()}
                                                                                  
    def __iter__(self):
        worker_info = torch.utils.data.get_worker_info()
        if worker_info is None:
            iter_start = self.start
            iter_end   = self.end
        else:  # multiple workers
            per_worker = int(np.ceil((self.end - self.start) / float(worker_info.num_workers)))
            worker_id = worker_info.id
            iter_start = self.start + worker_id * per_worker
            iter_end = min(iter_start + per_worker, self.end)
        sample_iterator = self._sample_generator(iter_start, iter_end)
        return sample_iterator
                                                                            
    def __len__(self):
        return self.end - self.start                                            
    
    def _get_file_info(self, file_path):
        end = 0
        time_slot_size = None                                                                 
        with open(file_path, 'r') as fin:                                       
            for _, line in enumerate(fin):
                if  time_slot_size == None:
                    _, _, _, _, _, label_mask = line.strip().split('|')
                    time_slot_size = len(label_mask.split(";")[0].split(":"))
                end += 1                                                                                            
        return time_slot_size, end                                                             
                                                                                
    def _sample_generator(self, start, end):
        with open(self.file_path, 'r') as fin:
            for i, line in enumerate(fin):
                if i < start:
                    continue                                          
                if i >= end:
                    return StopIteration()
                items = line.strip().split('|')
                first_layer_idx, day, start_time_slot, input_data, output_data, label_mask = items
                first_layer_idx = int(first_layer_idx)
                day = int(day)
                start_time_slot = int(start_time_slot)
                input_data = [[[float(kk) for kk in k.split(",")] for k in i_d.split(":")] for i_d in input_data.split(";")]
                output_data = [[[float(kk) for kk in k.split(",")] for k in i_d.split(":")] for i_d in output_data.split(";")]
                label_mask = [[int(k) for k in i_d.split(":")] for i_d in label_mask.split(";")]
                assert len(input_data) == len(output_data)
                assert len(output_data) == len(label_mask)
                date = datetime.datetime.strptime(str(day), "%Y%m%d")
                nxt_date = date + datetime.timedelta(days=1)
                weeks  = [date.weekday() if (start_time_slot+i)//288 == 0 else nxt_date.weekday()  for i in range(self.time_slot_size) ]
                minutes = [(start_time_slot+i) % 288  for i in range(self.time_slot_size)]
                edges = self.second_layer_edges[first_layer_idx]
                num_nodes = len(edges)
                u = []
                v = []
                for node_i, edge in enumerate(edges):
                    for e in edge:
                        u.append(node_i)
                        v.append(e)
                g = dgl.graph((u,v),num_nodes=num_nodes)
                node_idx = [self.new_part2id_dict[(first_layer_idx, s)] for s in range(num_nodes)]
                node_features = [self.new2old_edge_dict[i][1:3] for i in node_idx]
                g.ndata['f'] = torch.FloatTensor(node_features)
                yield { "first_layer_idx":first_layer_idx, 
                        "weeks": weeks,
                        "minutes": minutes,
                        "graph": g,
                        "x": input_data,
                        "mask": label_mask,
                        "y": output_data }
```

```python
#验证集查看结果
import tqdm
import torch.nn as nn
import numpy as np
device="cpu"
def validation(model, criterion, loader):
    model.eval()
    val_loss_1, val_loss_2 = [], []
    with torch.no_grad():
        for data in loader:
            first_layer_idx, weeks, minutes, batch_g, y, mask = data
            first_layer_idx, weeks, minutes, batch_g, y, mask = \
            first_layer_idx.to(device, non_blocking=True), weeks.to(device, non_blocking=True), \
                minutes.to(device, non_blocking=True), batch_g.to(device), \
                    y.to(device, non_blocking=True), mask.to(device, non_blocking=True)
            _, predict_1 = model(x=(batch_g, weeks, minutes, first_layer_idx),first_round=True)
            _, predict_2 = model(x=predict_1,first_round=False)
            loss_1 = criterion(predict_1, y, mask, reduction=True)
            loss_2 = criterion(predict_2, y, mask, reduction=True)
            # print(loss_1.shape, loss_2.shape)
            val_loss_1.append(loss_1.numpy())
            val_loss_2.append(loss_2.numpy())
    # return np.mean(val_loss_1), np.mean(val_loss_2)
    return val_loss_1, val_loss_2
            # print(predict_1)
            # print(predict_2)
            # print(torch.mean(y)*1.0/torch.mean(mask.float()),loss_1.item(),loss_2.item())
            # break
    # return val_loss_1, val_loss_2
def my_collate_fn(batch):
    first_layer_idx = []
    weeks = []
    minutes = []
    # x = []
    mask = []
    y = []
    batch_g = []
    batch_idx = 0
    for data in batch:
        first_layer_idx.append(data['first_layer_idx'])
        weeks.append(data['weeks'])
        minutes.append(data['minutes'])
        mask += data['mask']
        y += data['y']
        g = data['graph']
        g.ndata['batch_idx'] = torch.LongTensor([batch_idx for _ in range(g.num_nodes())])
        batch_g.append(g)
    return torch.LongTensor(first_layer_idx), torch.LongTensor(weeks),\
        torch.LongTensor(minutes), dgl.batch(batch_g),\
            torch.FloatTensor(y), torch.LongTensor(mask)
def traffic_loss(input, target, mask, reduction=False):
    loss_function = nn.SmoothL1Loss(reduction='none')
    # loss_function = nn.MSELoss(reduction='none')
    loss = loss_function(input, target)
    loss = torch.mean(loss, dim=-1)
    loss = torch.masked_select(loss, mask>0)
    if reduction:
        # import pickle
        # pickle.dump(loss, open("./loss.ts", "wb"))
        # print("mean:", loss.shape, torch.mean(loss).shape)
        return torch.mean(loss)
    else:
        # print("loss:", loss.shape)
        return loss
```
